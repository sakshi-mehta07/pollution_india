## Webscraping wikipedia

```python
#https://en.wikipedia.org/wiki/List_of_power_stations_in_India
```

```python
link_web = "https://en.wikipedia.org/wiki/List_of_power_stations_in_India"
cities_web = requests.get(link_web)
print(cities_web)
soup = BeautifulSoup(cities_web.content,"lxml")
```

```python
#https://scrapfly.io/blog/how-to-scrape-tables-with-beautifulsoup/

# Table 2,3,4 are coal, gas plants, those are the ones we want.
def tables_to_df(soup=soup,table_num=2):
    table = soup.find_all('table')[table_num]
    
    headers = []
    rows = []
    for i, row in enumerate(table.find_all('tr')):
        if i == 0:
            headers = [el.text.strip() for el in row.find_all('th')]
        else:
            rows.append([el.text.strip() for el in row.find_all('td')])

    return pd.DataFrame(rows,columns=headers)
```

```python
df_coal = tables_to_df(soup,table_num=2)
df_coal
```

```python
df_gas = tables_to_df(soup,table_num=3)
df_gas
```

```python
# Change directory structure here. TODO.
df_coal.to_csv("coal_plants_india_raw.csv")
```

```python
df_gas.to_csv("gas_plants_india_raw.csv")
```

### Coal plants

```python
df_coal_clean = pd.read_csv("DATA_INDIA/power_plants/coal_plants_india_raw.csv")
```

```python
df_coal_clean = df_coal_clean[["Location","District","State","Capacity(MW)","Sector"]]
```

```python
df_coal_clean = df_coal_clean.dropna(how="all")
```

```python
# Drop coal plants where capacity is not specified
df_coal_clean = df_coal_clean.dropna(how="all")
df_coal_clean = df_coal_clean.dropna(how="any", subset=["Capacity(MW)"])
```

```python
# Fixing some formatting from the original table. 
# Commas can be removed dots remain for the decimal! Do not replace aiaiai.

df_coal_clean['Capacity(MW)'] = df_coal_clean['Capacity(MW)'].astype("str").str.replace(",",""
                                                 ).str.replace("-","0"
                                                              ).str.replace("[22]","").astype("float")
```
